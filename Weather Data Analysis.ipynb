### Average value of each column against each Weather Condition

```python
numerical_columns = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%','Wind Speed_km/h', 'Visibility_km', 'Press_kPa']
df.groupby('Weather Condition')[numerical_columns].mean()
```

### Maximum value of each column against each Weather Condition

```python
numerical_columns = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%','Wind Speed_km/h', 'Visibility_km', 'Press_kPa']
df.groupby('Weather Condition')[numerical_columns].max()
```

### Minimum value of each column against each Weather Condition

```python
numerical_columns = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%','Wind Speed_km/h', 'Visibility_km', 'Press_kPa']
df.groupby('Weather Condition')[numerical_columns].min()
```

### All the records where 'Weather Condition' is Foggy or similar

### All instances when Weather is 'Clear' or Visibility is above 40

```python
df[(df['Weather Condition'] == 'Clear') | (df['Visibility_km'] > 40)]
```

### All the instances when: 'Weather' is 'Clear' and 'Relative Humidity' is greater than 50 or 'Visibility' is above 40

```python
df[(df['Weather Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
```
